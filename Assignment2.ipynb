# 문제 1
파일에서 읽은 영상에 대해 가로 세로를 1/3로 줄인 영상을 생성하는 함수를 작성함. Astronaut 영상을 이용하여 결과를 확인


````python
def reshape_AThird(img):
    r,c = img.shape[:2]
    new_img = cv2.resize(img, (int(r/3), int(c/3)))
    return new_img
````

````python
new_img = reshape_AThird(img)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
````

# 문제 2
파일에서 읽은 영상에 대해 다음과 같이 생성하는 함수를 작성함. Astronaut 영상(그레이로 변환된 영상)을 이용하여 결과를 확인.


* 원본을 그레이로 변환 

````python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
````

* 주어진 함수로 변환 해서 결과확인.

````python
def minimg(img):
    for y in range(len(img)):
        for x in range(len(img[0])):
            if img[y][x]<100:
                img[y][x] = 100
    return img
new_img = minimg(gray)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
````
