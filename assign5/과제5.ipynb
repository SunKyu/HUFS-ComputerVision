```python
import cv2 as cv
from matplotlib import pyplot as plt
```

# 실습1 


```python
gray = cv.imread('pattern.png')
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE,(15, 15))
eroding = cv.erode(gray, kernel, iterations = 1)
opening = cv.morphologyEx(gray, cv.MORPH_OPEN, kernel)
dilation = cv.dilate(gray, kernel, iterations = 1)
closing = cv.morphologyEx(gray, cv.MORPH_CLOSE, kernel)
```

## 원본

```python
plt.imshow(cv.cvtColor(gray, cv.COLOR_BGR2RGB))
```

## ⊖⊗⊕⊝∘≢⋅∘

## A⊖B

```python
plt.imshow(eroding)
```

## A∘B = (A⊖B) ⊕B

```python
plt.imshow(opening)
```

## A⊕B

```python
plt.imshow(dilation)
```

## A⋅B = (A⊕B)⊝B

```python
plt.imshow(closing)
```

# 실습 3

```python
img = cv.imread('0911.png')
kernel = cv.getStructuringElement(cv.MORPH_RECT,(3, 3))
plt.imshow(img)
```

```python
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.imshow(opening)
```

```python
res = cv.morphologyEx(opening, cv.MORPH_CLOSE, kernel)
plt.imshow(res)
```
