```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def color_rect():
    g = np.zeros((256,256,3), np.uint8)
    
    for i in range(256):
        g[i,:,:] = (0, i, 255-i)
    
    return g
```

```python
g = color_rect()
plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
```

```python

f1 = [255,0, 0]
f2 = [0, 0, 255]
f3 = [0, 255, 0]
f4 = [255, 255, 0]
def color_rect2():
    g = np.zeros((256,256,3), np.uint8)
    for i in range(256):
        for j in range(256):
            lis = []
            for k in range(3):
                x = np.array(([255-i, i]))
                y = np.array(([255-j],[j]))
                far = np.array(([f1[k], f2[k]], [f3[k], f4[k]]))
                z = np.matmul(x,far)
                z = np.matmul(z,y)
                lis.append(z)
                
            g[i,j,:] = lis
    return g
```

```python
g = color_rect2()
print(g[0,0])
print(g[0,255])
print(g[255,0])
print(g[255,255])


plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
```

```python
def color_rect3():
    g = np.zeros((256,256,3), np.uint8)
    
    for i in range(256):
        for j in range(256):
            g[i,j,:] = (255-i-j, i, j)
    
    return g
```

```python
g = color_rect3()
print(g[0,0])
print(g[0,255])
print(g[255,0])
print(g[255,255])


plt.imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
```
